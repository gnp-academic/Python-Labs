# Exceptions: try/except, else, finally; Common Errors; raise

**Exercise 1:** Wrap a division `a/b` in try/except to catch `ZeroDivisionError` and print a helpful message.

**Exercise 2:** Convert the string `'42'` to int inside try/except/else/finally; print success in else, 'done' in finally.

**Exercise 3:** Open a missing file path and catch `FileNotFoundError` gracefully; print a friendly message.

**Exercise 4:** Use `raise` to throw a `ValueError` with a custom message if a number is negative; test with `-1` and catch it.

**Exercise 5:** Handle multiple exceptions (`ValueError` and `TypeError`) when converting `x` to int for `x in ['15','oops',None]`.

**Exercise 6:** Define a custom exception class `TooSmallError`. Write a function that raises it if input < 10; test with 7 and handle it.

**Exercise 7:** Read an integer from a string; if conversion succeeds (else branch), print its square; otherwise print 'bad input'. Use `s='16'`.

**Exercise 8:** Demonstrate logging + re-raise: in a try block, divide by zero; in except, print 'logging...' then re-raise. Catch it in an outer try to prevent crash.

**Exercise 9:** Use `assert` to ensure a list is non-empty; catch the resulting `AssertionError` and print a message.

**Exercise 10:** Simulate resource acquisition with a variable `opened=True`; in try, raise `RuntimeError`; in finally, set `opened=False` and print its value.
